```python
import pandas as pd
import matplotlib.pyplot as plt


bmi = data['BMI']
difficulty_walking = data['DiffWalking']


filtered_data = data[(bmi < 25) & (difficulty_walking == 'Yes')]

count_with_difficulty_walking = len(filtered_data)
count_without_difficulty_walking = len(data[(bmi < 25) & (difficulty_walking == 'No')])

labels = ['With DiffWalking', 'Without DiffWalking']
sizes = [count_with_difficulty_walking, count_without_difficulty_walking]
colors = ['lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('BMI < 25 and DiffWalking')

plt.axis('equal')
plt.show()
```

```python
sleep_pct = []
sleep_time=[]
for i in range(1,25):
    sleep_time.append(float(i))
    heart_diseased = len(data[(data["SleepTime"]==float(i)) &(data["HeartDisease"]=="Yes")])
    total_shleep = len(data[data["SleepTime"]==float(i)])
    sleep_pct.append(heart_diseased/total_shleep)

plt.scatter(sleep_time,sleep_pct)
plt.show()
```

```python
sleep_df = pd.DataFrame({"SleepTime":sleep_time,"Sleep_Percent":sleep_pct})
sleep_df
```
